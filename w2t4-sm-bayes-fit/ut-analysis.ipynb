## *and determination of the ckm parameters*

In the SM flavor changing charged currents happen through W couplings to quark.

<center>

</center>

These couplings are defined in the CKM matrix:

$$
V_{CKM} = \begin{pmatrix}{}
V_{ud} & V_{us} &  V_{ub} \\
V_{cd} & V_{cs} &  V_{cb} \\
V_{td} & V_{ts} &  V_{tb}
\end{pmatrix}
$$

To a good level of precision $V_{CKM}$ can be written as:

$$
V_{CKM} = \begin{pmatrix}{}
1 - \frac{\lambda^2}{2} & \lambda &  A\lambda^3 (\bar{\rho} - i \bar{\eta}) \\
- \lambda & 1 - \frac{\lambda^2}{2} &  A\lambda^2 \\
A\lambda^3 (1 - \bar{\rho} - i \bar{\eta}) & -A\lambda^2 & 1
\end{pmatrix}
+ O(\lambda^8)
$$

where $\lambda \simeq 0.22$ is $\sin(\theta_c)$ ($\theta_c$ the Cabibbo angle).

The expression above involves 4 parameters: $(A, \lambda, \bar{\rho}, \bar{\eta})$.
$\bar{\eta} \neq 0$ yields CP violation.

CKM parameters can be measured in several ways:

### Semileptonic decays of the $B$ meson

<center>
    
</center>

$$
BR(B^+ \to \ell^+ \nu \bar{D}^0) \sim \left|V_{cb}\right|^2
$$

<center>
    
</center>

$$
BR(B^+ \to \ell^+ \nu \pi^0) \sim \left|V_{ub}\right|^2
$$

These are expressed by the above parametrization as:

$$
\begin{align}
|V_{ub}| = A \lambda^3 \sqrt{\bar{\rho}^2 + \bar{\eta}^2} \\
|V_{cb}| = A \lambda^2 \\
\left|\frac{V_{ub}}{V_{cb}}\right| \propto \sqrt{\bar{\rho}^2 + \bar{\eta}^2}
\end{align}
$$

<center>
    
</center>

### CP violation in $B \to J/\Psi K_s$

<center>

</center>

$$
A_{CP} = \frac{
N(B^0 \to J/\Psi K_s) - N(\bar{B}^0 \to J/\Psi K_s)
}{
N(B^0 \to J/\Psi K_s) + N(\bar{B}^0 \to J/\Psi K_s)
} \sim \sin(2 \beta)
$$

<center>

</center>

$$
\beta = \tan^{-1} \left(\frac{\bar{\eta}}{1 - \bar{\rho}}\right)
$$

Moreover, similar experiments can measure:
$$
\gamma = \tan^{-1} \left(\frac{\bar{\eta}}{\bar{\rho}}\right)
$$
and
$$
\alpha = \pi - \gamma - \beta
$$
